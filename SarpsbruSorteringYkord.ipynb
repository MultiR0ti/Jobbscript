```python
#sort on y
with arcpy.da.UpdateCursor(fc_in, uc_fields) as uc:
    for row in uc:
        if row[2] == 'Totalsondering' and s_y[row[0]]<65:
            if s_y[row[0]]<10:
                row[1] = 'SB-100'+str(s_y[row[0]])
                uc.updateRow(row)
            else:
                row[1] = 'SB-10'+str(s_y[row[0]])
                uc.updateRow(row)
        elif row[2] == 'Totalsondering' and s_y[row[0]]>64:
            if s_y[row[0]]-64<10:
                row[1] = 'SB-200'+str(s_y[row[0]]-64)
                uc.updateRow(row)
            else:
                row[1] = 'SB-20'+str(s_y[row[0]]-64)
                uc.updateRow(row)
        else:    
            row[1] = ''
            uc.updateRow(row)
        print(row)
```
